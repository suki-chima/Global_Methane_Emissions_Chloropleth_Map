```python
import plotly.express as px

# Assuming you have a DataFrame called 'df3' with columns "country", "emissions", and "year"

# Define the dropdown options
dropdown_options = [
    {'label': 'Fig1 (2022)', 'value': 'fig1'},
    {'label': 'Fig2 (2019-2021)', 'value': 'fig2'}
]

custom_color_scale = [
    (0.0, 'green'),   # For the minimum value (0.0), use green color
    (0.2, 'yellow'),   # For the minimum value (0.2), use yellow color
    (0.5, 'orange'),  # For the middle value (0.5), use orange color
    (1.0, 'red')     # For the maximum value (1.0), use red color
]


# Assuming you want to create a choropleth map for the year
selected_year1 = '2022'
df4= df3.loc[df3['year'] == selected_year1]

selected_year2 = '2019-2021'
df5= df3.loc[df3['year'] == selected_year2]

# Create the choropleth map for the year 2022 using Plotly Express
fig1 = px.choropleth(
    data_frame=df4,
    locations="country",
    locationmode='country names',
    color="emissions",
    color_continuous_scale=custom_color_scale,
    projection="natural earth"
)

# Create the choropleth map for the years 2019 to 2021 using Plotly Express
fig2 = px.choropleth(
    data_frame=df5,
    locations="country",
    locationmode='country names',
    color="emissions",
    color_continuous_scale=custom_color_scale,
    projection="natural earth"
)

# Set the title for each choropleth map
fig1.update_layout(title="IEA estimates for methane emissions (kiotonnes) by country in 2022")
fig2.update_layout(title="IEA estimates for methane emissions (kiotonnes) by country in 2019-2021")

# Display the dropdown menu to select the figure
print("Select a figure to display:")
for index, option in enumerate(dropdown_options, 1):
    print(f"{index}. {option['label']}")

selected_value = None
while selected_value not in ('fig1', 'fig2'):
    try:
        selected_index = int(input("Enter the figure number: "))
        selected_value = dropdown_options[selected_index - 1]['value']
    except (ValueError, IndexError):
        print("Invalid input. Please enter a valid figure number.")

# Display the selected figure
if selected_value == 'fig1':
    fig1.show()
else:
    fig2.show()
```
